```python
import numpy as np

def one_hot_encode(df):
    
    sizes = [3, 3, 2, 3, 4, 2]

    X = df[['A1','A2','A3','A4','A5','A6']].values
    encoded = []

    for i, size in enumerate(sizes):
        n_samples = X.shape[0] # number of samples
        print(n_samples)
        n_classes = size # number of unique values in the column
        temp = np.zeros((n_samples, n_classes))
        
        row_indices = np.arange(n_samples) 
        col_indices = X[:, i] - 1  # assuming values start from 1 , 0 based indexing
        temp[row_indices, col_indices] = 1
        encoded.append(temp)

    return np.hstack(encoded)
one_hot_encode(monks2_train_df)
```

**Network architecture**

```python
# Network architecture
def init_weights(input_size, hidden_size, output_size, activation='sigmoid',init_type='xavier', seed=None):
   if seed is not None:
       np.random.seed(seed)
       
   if init_type == 'small_random':
      weights_hidden_output = np.random.randn(hidden_size, output_size) * 0.01
      weights_input_hidden = np.random.randn(input_size, hidden_size) * 0.01
   elif init_type == 'xavier':
      scale_input_hidden = np.sqrt(1 / input_size)
      scale_hidden_output = np.sqrt(1 / hidden_size)
      weights_input_hidden = np.random.randn(input_size, hidden_size) * scale_input_hidden
      weights_hidden_output = np.random.randn(hidden_size, output_size) * scale_hidden_output
   elif init_type == 'he':
      scale_input_hidden = np.sqrt(2 / input_size)
      scale_hidden_output = np.sqrt(2 / hidden_size)
      weights_input_hidden = np.random.randn(input_size, hidden_size) * scale_input_hidden
      weights_hidden_output = np.random.randn(hidden_size, output_size) * scale_hidden_output
   else:
      raise ValueError("Unknown init_type. Use 'small_random', 'xavier', or 'he'.")
   
   bias_hidden = np.zeros((1, hidden_size))
   bias_output = np.zeros((1, output_size))
   
   return weights_input_hidden, weights_hidden_output, bias_hidden, bias_output
```

**Activation Functions**

```python
def relu(x):
    """ReLU activation function"""
    return np.maximum(0, x)

def relu_derivative(x):
    """Derivative of ReLU"""
    return (x > 0).astype(float)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)
```

```python
def tanh(x): return np.tanh(x)
def tanh_deriv(x): return 1 - np.tanh(x)**2
```

**Forward Pass**

```python
def forward_pass(X, weights_input_hidden, weights_hidden_output, bias_hidden, bias_output, activation='sigmoid'):
    # Hidden layer
    z_hidden = np.dot(X, weights_input_hidden) + bias_hidden
    if activation == 'sigmoid':
        a_hidden = sigmoid(z_hidden)
    else:
        a_hidden = tanh(z_hidden)
    
    # Output layer
    z_output = np.dot(a_hidden, weights_hidden_output) + bias_output
    a_output = sigmoid(z_output)
    
    return a_hidden, a_output
```

**Compute Loss**

```python
def compute_loss(y_true, y_pred):
    m = y_true.shape[0]
    loss = - (1/m) * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return loss
```

**Backpropagation**

```python
def backward_pass(X, y, a_hidden, a_output,weights_input_hidden,bias_hidden, weights_hidden_output, bias_output,learning_rate=0.01,activation='sigmoid'):
   #  print("X:", X.shape)
    m = y.shape[0] # number of samples
    
    # Output layer gradients
    delta_output = a_output - y.reshape(-1, 1) 
    grad_weights_hidden_output = (1/m) * np.dot(a_hidden.T, delta_output)
    grad_bias_output = (1/m) * np.sum(delta_output, axis=0, keepdims=True)
    
    # Hidden layer gradients
    if activation == 'sigmoid':
        delta_hidden = np.dot(delta_output, weights_hidden_output.T) * (a_hidden * (1 - a_hidden)) 
    else:
        delta_hidden = np.dot(delta_output, weights_hidden_output.T) * tanh_deriv(a_hidden) 

    grad_weights_input_hidden = (1/m) * np.dot(X.T, delta_hidden) 
    grad_bias_hidden = (1/m) * np.sum(delta_hidden, axis=0, keepdims=True)
    
    # Update weights and biases
        
    weights_hidden_output -= learning_rate * grad_weights_hidden_output
    bias_output -= learning_rate * grad_bias_output
    weights_input_hidden -= learning_rate * grad_weights_input_hidden
    bias_hidden -= learning_rate * grad_bias_hidden
    
    return grad_weights_input_hidden, grad_bias_hidden, grad_weights_hidden_output, grad_bias_output
```

**Training Loop**

```python
def train_monk(X_train_monks2, y_train_monks2, X_test_monks2, y_test_monks2, hidden_size,epochs,learning_rate, activation='sigmoid', init_type='xavier',momentum=0.9, early_stop=True, patience=500 ,seed=None):

    input_size = X_train_monks2.shape[1]
    output_size = 1
    
    # Initialize weights
    weights_input_hidden, weights_hidden_output, bias_hidden, bias_output = init_weights(input_size, hidden_size, output_size, activation, init_type, seed)
    
    #Initialize momentum terms
    v_weights_input_hidden = np.zeros_like(weights_input_hidden)
    v_bias_hidden = np.zeros_like(bias_hidden)
    v_weights_hidden_output = np.zeros_like(weights_hidden_output)
    v_bias_output = np.zeros_like(bias_output)
    
    
    loss_history = []
    train_acc_history = []
    test_acc_history = []

    #Initialize Early Stopping variables
    best_test_accuracy = 0.0
    best_epoch = 0
    
    best_weights_input_hidden = weights_input_hidden.copy()
    best_weights_hidden_output = weights_hidden_output.copy()
    best_bias_hidden = bias_hidden.copy()
    best_bias_output = bias_output.copy()
    
    count_patience = 0
    
    for epoch in range(epochs):
        a_hidden, a_output = forward_pass(X_train_monks2, weights_input_hidden, weights_hidden_output, bias_hidden, bias_output, activation)

        # Compute loss
        loss = compute_loss(y_train_monks2.reshape(-1, 1), a_output)
        
        #Record loss
        loss_history.append(loss)
        
        # Record training accuracy
        y_pred_classes = (a_output > 0.5).astype(int).flatten()
        train_accuracy = np.mean(y_pred_classes == y_train_monks2)  
        train_acc_history.append(train_accuracy)
        
        # Record test accuracy
        _, a_output_test = forward_pass(X_test_monks2, weights_input_hidden, weights_hidden_output, bias_hidden, bias_output, activation) 
        y_pred_test_classes = (a_output_test > 0.5).astype(int).flatten()
        test_accuracy = np.mean(y_pred_test_classes == y_test_monks2)
        test_acc_history.append(test_accuracy)
        
        # Early Stopping check
        if test_accuracy > best_test_accuracy:
            best_test_accuracy = test_accuracy
            best_epoch = epoch
            
            best_weights_input_hidden = weights_input_hidden.copy()
            best_weights_hidden_output = weights_hidden_output.copy()
            best_bias_hidden = bias_hidden.copy()
            best_bias_output = bias_output.copy()
            
            count_patience = 0
        else:
            count_patience += 1 
        
        if count_patience >= patience:
            print(f'Early stopping at epoch {epoch}. Best test accuracy: {best_test_accuracy*100:.2f}% at epoch {best_epoch}.')
            break
        
        # Backward pass
        grad_weights_input_hidden, grad_bias_hidden, grad_weights_hidden_output, grad_bias_output = backward_pass(
        X_train_monks2, y_train_monks2, a_hidden, a_output,
        weights_input_hidden, bias_hidden,
        weights_hidden_output, bias_output,
        learning_rate=learning_rate,
        activation='sigmoid')
        
        #Apply momentum
        v_weights_input_hidden = momentum * v_weights_input_hidden - learning_rate * grad_weights_input_hidden
        v_bias_hidden = momentum * v_bias_hidden - learning_rate * grad_bias_hidden
        v_weights_hidden_output = momentum * v_weights_hidden_output - learning_rate * grad_weights_hidden_output
        v_bias_output = momentum * v_bias_output - learning_rate * grad_bias_output
        
        # Update weights with momentum
        weights_input_hidden += v_weights_input_hidden
        bias_hidden += v_bias_hidden
        weights_hidden_output += v_weights_hidden_output
        bias_output += v_bias_output
        
        # Print loss every 10 epochs
        if epoch % 100== 0:
            print(f'Epoch {epoch}, Loss: {loss:.4f}, Train Acc: {train_accuracy*100:.2f}%, Test Acc: {test_accuracy*100:.2f}%')
    
    if early_stop:
        weights_input_hidden = best_weights_input_hidden
        weights_hidden_output = best_weights_hidden_output
        bias_hidden = best_bias_hidden
        bias_output = best_bias_output
        
        print(f"Best model at epoch {best_epoch} with Test Acc: {best_test_accuracy*100:.2f}%")        
    return weights_input_hidden, weights_hidden_output, bias_hidden, bias_output, loss_history, train_acc_history, test_acc_history   
```

```python
import matplotlib.pyplot as plt

def plot_results(loss_history, train_acc_history, test_acc_history):
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(loss_history, label='Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss') 
    plt.title('Training Loss')
    
    plt.subplot(1, 2, 2)
    plt.plot(train_acc_history, label='Training Accuracy')
    plt.plot(test_acc_history, label='Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Training/Test Accuracy')
    plt.legend()
    
    plt.show()
```

```python
# Train MONK-1 with different configurations
weights_input_hidden, bias_hidden, weights_hidden_output, bias_output, loss_hist, train_acc, test_acc = train_monk(
    X_train_monks2, y_train_monks2,
    X_test_monks2, y_test_monks2,
    hidden_size=4,
    epochs=1000,
    learning_rate=0.4,
    activation='sigmoid',  # or 'tanh'
    init_type='xavier',    # 'small_random', 'xavier', 'he'
    momentum=0.9,
    seed=42
)   

plot_results(loss_hist, train_acc, test_acc)
```
